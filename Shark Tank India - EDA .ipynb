Q2. Max, min, mean of asked amount, asked equity, asked valuation? 

```python
#Max, min, mean of asked amount
max_asked_amount = data['pitcher_ask_amount'].max()
min_asked_amount = data['pitcher_ask_amount'].min()
mean_asked_amount = data['pitcher_ask_amount'].mean()

#Max, min, mean of asked equity
max_asked_equity = data['ask_equity'].max()
min_asked_equity = data['ask_equity'].min()
mean_asked_equity = data['ask_equity'].mean()

#Max, min, mean of asked valuation
max_asked_valuation = data['ask_valuation'].max()
min_asked_valuation = data['ask_valuation'].min()
mean_asked_valuation = data['ask_valuation'].mean()

print("Max Asked Amount:", max_asked_amount)
print("Min Asked Amount:", min_asked_amount)
print("Mean Asked Amount:", mean_asked_amount)

print("Max Asked Equity:", max_asked_equity)
print("Min Asked Equity:", min_asked_equity)
print("Mean Asked Equity:", mean_asked_equity)

print("Max Asked Valuation:", max_asked_valuation)
print("Min Asked Valuation:", min_asked_valuation)
print("Mean Asked Valuation:", mean_asked_valuation)
```

Q10. How many sharks participated in this show and What were their names? 

```python
from wordcloud import WordCloud
having = 'present'
count = 0
matching_columns = [col for col in data.columns if having in col]
names = ""
for col in matching_columns:
    shark_names = col.replace('_present', '')
    names += shark_names + "\n"
    count += 1

print("Total number of sharks are" , count)
print("Names of Sharks : " , names)
# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(names)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Q11. Find appearance of each sharks? 

```python
having = 'present'
matching_columns = [col for col in data.columns if having in col]
names = ''
for col in matching_columns:
    names += col + "\n"
array = names.split('\n')
array = [item for item in array if item]  # Filter out empty strings
ab=data[array]
Present=ab.sum().sort_values()
print (Present)
plt.figure(figsize=(10, 6))
Present.plot(kind='bar', color='dodgerblue')
plt.title('Number of Appearances for Each Shark')
plt.xlabel('Shark Names')
plt.ylabel('Number of Appearances')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Q13. How many times each shark invested the deal?

Q14. Find the equity percent that each sharks gets? 

```python
having = '_deal'
matching_columns = [col for col in data.columns if having in col]
names = ''
l = []
for col in matching_columns:
    names += col + "\n"
array = names.split('\n')
array = [item for item in array if item]  # Filter out empty strings
ab=data[array]
for i in ab:
    x=data[data[i]==1]
    s=x["equity_per_shark"].sum()
    l.append(s)
z = sum(l)
equity_percentages = []
for i in ab:
    x=data[data[i]==1]
    s=x["equity_per_shark"].sum()
    l.append(s)
    print(i,"- total equity -",s ," - ", s*100/z , "%")
    percent = s*100/z
    equity_percentages.append(percent)
plt.figure(figsize=(10, 6))
plt.bar(array, equity_percentages, color='dodgerblue')
plt.title('Equity Percentage for Each Shark')
plt.xlabel('Shark Names')
plt.ylabel('Equity Percentage (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
